```python
import numpy as np
import matplotlib.pyplot as plt

# x değerleri üretilir
x = np.linspace(0, 10, 50)

# Doğrusal bir ilişki oluşturun (hatasız)
true_y = 2 * x + 1

# Normal dağıtılmış hatalar üret
errors = np.random.normal(0, 1, 50)

# Gürültülü veriler oluşturmak için gerçek değerlere hatalar ekleyin
y = true_y + errors

# Gürültülü veri noktalarının dağılım grafiğini oluşturun
plt.scatter(x, y, label='Noisy Data', color='b', alpha=0.7)

# Gerçek eğrinin bir grafiğini oluşturun (hatasız)
plt.plot(x, true_y, label='True Curve', color='r', linestyle='--')

# Etiketleri ve açıklamaları ayarlayın
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Göster
plt.show()

# Formülü yazdır
print("Linear Regression Formula: y = 2x + 1")
```
